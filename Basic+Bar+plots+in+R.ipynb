#### Lets use in-built mtcars data to make some basic bar plots in R

```r
head(mtcars)
```

#### Lets check how many factor the gear column have

```r
Gear <- table(mtcars$gear)
Gear
```

## Bar Plots

```r
# Plot the gear column
barplot(Gear)
```

The above thing can be achieved simply by using ggplot2 package of R. Lets first use qplot of ggplot2. 

```r
library(ggplot2)
qplot(mtcars$gear, geom = 'bar')
```

To fill the color, making border line of each bar and renaming the axis do the following.

```r
qplot(mtcars$gear, geom = 'bar',  color= I('blue'), fill = I('yellow'), xlab = 'Gear', ylab = 'Count of Vehicles', main = 'Gear in mtcars')
```

Now use ggplot function to do the same thing

```r
ggplot(aes(gear), data = mtcars) + geom_bar()
```

```r
ggplot(aes(gear), data = mtcars) + geom_bar(fill='#A4A4A4', color="darkred") + ylab('Count of Vehicle') + xlab('Gear') + ggtitle('Plot of Gear') 
```
